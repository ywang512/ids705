```python
dict_id_borough = pd.read_csv("./taxi+_zone_lookup.csv", index_col=0)["Borough"].to_dict()
```

```python
data = pd.read_csv("./fhv_tripdata_2018-01.csv")
```

```python
data.head(20)
```

```python
from collections import defaultdict
PU_count = defaultdict(int)
DO_count = defaultdict(int)
```

```python
for i in range(len(data)):
    if i % 100000 == 0:
        print("=", end="")
    if i % 1980809 == 0:
        print(">")
    PU_id = data.iloc[i, 2]
    DO_id = data.iloc[i, 3]
    if not pd.isna(PU_id) and PU_id != 0:
        PU_count[dict_id_borough[PU_id]] += 1
    if not pd.isna(DO_id) and DO_id != 0:
        DO_count[dict_id_borough[DO_id]] += 1
```

```python
PU_count
```

```python
DO_count
```

```python
a, b = zip(*PU_count.items())
plt.figure(figsize=(8,6))
plt.bar(a, b)
plt.title("PickUp Location Counts")
plt.show()
```

```python
a, b = zip(*DO_count.items())
plt.figure(figsize=(8,6))
plt.bar(a, b)
plt.title("DropOff Location Count")
plt.show()
```
